```python
category = df['Category'].unique()
print(category)
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
category_mean_air_quality = df.filter(['AQI','Category', 'population', 'density'])
category_mean_air_quality = category_mean_air_quality.groupby('Category').mean().reset_index()
category_mean_air_quality = category_mean_air_quality.sort_values('AQI').reset_index()
category_mean_air_quality.head(6)
```

```python
category_mean_air_quality_state = df.filter(['Date','AQI', 'state_name', 'Category', 'population', 'density'])
category_mean_air_quality_state['month'] = pd.DatetimeIndex(category_mean_air_quality_state['Date']).month
category_mean_air_quality_state['year'] = pd.DatetimeIndex(category_mean_air_quality_state['Date']).year
category_mean_air_quality_state.head()
```

```python
category_mean_air_quality_state2 = category_mean_air_quality_state.groupby(['state_name','year','month']).mean().reset_index()
category_mean_air_quality_state2 = category_mean_air_quality_state2.sort_values(['state_name','year','month']).reset_index()
category_mean_air_quality_state2.head(20)
```
